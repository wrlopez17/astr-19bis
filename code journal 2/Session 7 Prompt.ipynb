```python
import matplotlib.pyplot as plt
import numpy as np

# Step 1: Create an array x ranging from [0,1] in 100 steps
x = np.linspace(0, 1, 100)
```

```python
# Step 2: Define a function that returns exp(x)
def exp_function(x):
    return np.exp(x)
```

```python
# Using the function to set y=exp(x)
y = exp_function(x)

# Plot x vs. y
plt.figure()
plt.plot(x, y)
plt.xlabel('Time [milliseconds]')
plt.ylabel('Awesomeness')
plt.title('Plot of Exponential Function')

# Save the figure as a PDF
plt.savefig('exp_plot.pdf')
```
