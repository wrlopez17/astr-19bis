```python
import matplotlib.pyplot as plt
import numpy as np

# Create an array x ranging from [0,1] in 100 steps
x = np.linspace(0, 1, 100)
```

```python
# Define functions that return sin(x) and cos(x)
def sin_function(x):
    return np.sin(x)

def cos_function(x):
    return np.cos(x)
```

```python
# Create a multipanel plot (1 row, 2 columns)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Plot sin(x) vs. x in the left panel
ax1.plot(x, sin_function(x))
ax1.set_title('sin(x)')
ax1.set_xlabel('x')
ax1.set_ylabel('sin(x)')

# Plot cos(x) vs. x in the right panel
ax2.plot(x, cos_function(x))
ax2.set_title('cos(x)')
ax2.set_xlabel('x')
ax2.set_ylabel('cos(x)')

# Save the figure as a PDF in the current working directory
plt.savefig('sin_cos_plot.pdf')
```
